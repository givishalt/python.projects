```python
import matplotlib.pyplot as plt
import seaborn as sns

# Test score data
scores = [55, 57, 58, 60, 62, 65, 65, 68, 90, 95]

# Set the style
sns.set(style="whitegrid")

# Create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1️⃣ Histogram
sns.histplot(scores, bins=5, kde=False, ax=axes[0], color='skyblue')
axes[0].set_title('Histogram of Test Scores')
axes[0].set_xlabel('Score')
axes[0].set_ylabel('Count')

# 2️⃣ Box Plot
sns.boxplot(scores, ax=axes[1], color='lightgreen')
axes[1].set_title('Box Plot of Test Scores')
axes[1].set_xlabel('Score')

# 3️⃣ KDE / Density Plot
sns.kdeplot(scores, ax=axes[2], color='salmon', fill=True)
axes[2].set_title('KDE / Density Plot of Test Scores')
axes[2].set_xlabel('Score')

plt.tight_layout()
plt.show()
```
