```python
import pandas as pd
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
df=pd.read_csv('Boston1.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression #fit and predict
from sklearn.metrics import  r2_score , mean_squared_error , mean_absolute_error
```

```python
df.head()
```

```python
#select data for modeling
```

```python
df_selected= df[['rm', 'ptratio','lstat','medv']]
df_selected.head(2)
```

```python
# refer stats model selection
```

```python
x=df_selected[['rm', 'ptratio','lstat']]
y=df_selected[['medv']]
print(x.shape)
print(y.shape)
```

```python
# split the data into train and test data
x_train, x_test , y_train , y_test = train_test_split(x,y, test_size = 0.3 , random_state=42)
```

```python
#configuring the model
model=LinearRegression()
model
```

```python
#fit the model
lm=model.fit(x_train, y_train)
```

```python
lm.coef_   #b0,b1,b2
```

```python
lm.intercept_  #constant
```

```python
#predict the train data
y_train_predict=lm.predict(x_train)
y_train_predict[:10]
```

```python
#predict the test data
y_test_predict=lm.predict(x_test)
y_test_predict[:10]
```

```python
#evaluation the  model----r squared, mean squared error for train data
rsq = r2_score(y_train, y_train_predict)
print(f'r squared value for train data is { rsq } ')
mse= mean_squared_error(y_train , y_train_predict )
print(f'mean squared value for train data  is {mse}' )

# root mean squared 

rmse = np.sqrt(mse)
print(f'root mean squared error of train data is { rmse} ')
```

```python
# metrics for test data 
rsq=r2_score(y_test , y_test_predict)
print(f'r squared value for  the test data is {rsq}')
mse= mean_squared_error(y_test , y_test_predict)
print(f'mean squared value for test data is {mse}')
rmse=np.sqrt(mse)
print(f'root mean squared error of test data is {rmse} ')
```
