```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Days= [ "sun", "mon","tue","wed","thu","fri","sat"]
earning=[ 100 , 200 , 300, 150, 123,300,  123]

plt.plot( Days , earning , marker = "o",ms=20, mec = 'r', mfc='blue', linestyle = "solid" , color = "black", label="Earnings per day")

plt.xlabel("Days")
plt.ylabel("earning")
plt.title(" sales trends ")
plt.grid()
plt.legend(loc="upper left",labelspacing=0.5, handlelength=1, fontsize= 8, handleheight=0.8)
plt.show()

# Parameter	Full Form	Meaning / Controls	Example Value	Description
# ms -	Marker Size	Size of marker symbols	20	Makes markers larger or smaller
# mec -	Marker Edge Color	Color of the border of markers	'r'	Red outline around marker
# mfc -	Marker Face Color	Color of the inside (fill) of marker	'blue'	Blue fill inside marker

# plt.legend(["plot1","plot2"],loc="best") #if we are having two plot
```
