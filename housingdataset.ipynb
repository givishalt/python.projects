```python
df = pd.read_csv(r"C:\Users\lenovo\Downloads\Downloads\ToyotaCorolla.csv")
```

```python
#Use the ToyotaCorollaX.csv of size : 1436 x 11 for the questions mentioned below Choose the correct CC value having the second lowest price?
```

```python
df['FuelType'].fillna(df['FuelType'].mode()[0], inplace=True)

# Verify
print(df['FuelType'].isnull().sum())  # Should be 0
```

```python


# Sort the dataframe by Price
df_sorted = df.sort_values(by='Price')

# Get the second lowest price row
second_lowest = df_sorted.iloc[1]  # iloc[0] is lowest, iloc[1] is second lowest

# Display CC value and corresponding Price
print(f"Second lowest price: {second_lowest['Price']}")
print(f"CC value at second lowest price: {second_lowest['CC']}")
```

```python
mean_price_by_doors = df.groupby('Doors')['Price'].mean().reset_index()

# Sort by mean price descending
mean_price_sorted = mean_price_by_doors.sort_values(by='Price', ascending=False)

# Get the second highest mean price
second_highest = mean_price_sorted.iloc[1]  # iloc[0] is highest, iloc[1] is second highest

print(f"Doors type with second highest mean price: {second_highest['Doors']}")
print(f"Mean price: {second_highest['Price']}")
```
