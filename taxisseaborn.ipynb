```python
plt.figure(figsize=(7,5))
sns.set_style('whitegrid')
sns.boxplot( data=df , y='fare' ,meanprops={
        "marker":"o",
        "markerfacecolor":"red",
        "markeredgecolor":"black"}, showmeans= True )
plt.ylabel('fare charged')
plt.title('boxplot for fares')
plt.show()



# there are more number of outliers here in fares column and all are greater than the upper fence 
```
