```python
'''1. Create a function named ‘factor’ that can only accept 1 argument. The function should return the factorial of that number.'''
def factor(n):
    fact = 1
    for i in range (1,n+1):
        fact *= i
    return fact

print(f"the factorial of the given number is {factor(6)}")
```
